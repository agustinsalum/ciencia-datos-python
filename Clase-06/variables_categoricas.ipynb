```python
# Importar librerías
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Crear un DataFrame de ejemplo con variables categóricas
data = {
    'Nombre': ['Ana', 'Juan', 'Pedro', 'Luisa', 'María', 'Carlos', 'Laura', 'Sofía', 'Diego', 'Lucía',
               'Miguel', 'Elena', 'Andrés', 'Valentina', 'Rubén', 'Isabel', 'David', 'Claudia', 'Felipe', 'Ana'],
    'Género': ['Femenino', 'Masculino', 'Masculino', 'Femenino', 'Femenino', 'Masculino', 'Femenino', 'Femenino', 'Masculino', 'Femenino',
               'Masculino', 'Femenino', 'Masculino', 'Femenino', 'Masculino', 'Femenino', 'Masculino', 'Femenino', 'Masculino', 'Femenino'],
    'Ciudad': ['Madrid', 'Barcelona', 'Valencia', 'Madrid', 'Barcelona', 'Valencia', 'Madrid', 'Barcelona', 'Valencia', 'Madrid',
               'Barcelona', 'Valencia', 'Madrid', 'Barcelona', 'Valencia', 'Madrid', 'Barcelona', 'Valencia', 'Madrid', 'Barcelona'],
    'Edad': [25, 30, 22, 27, 29, 32, 28, 26, 31, 24, 33, 35, 29, 26, 28, 27, 34, 25, 30, 33]
}

df = pd.DataFrame(data)
```

```python
# Mostrar el DataFrame
print("DataFrame:")
print(df)
```

```python
# Gráfico de barras para una variable categórica
plt.figure(figsize=(8, 6))
sns.countplot(x='Género', data=df)
plt.title('Distribución de Género')
plt.xlabel('Género')
plt.ylabel('Cantidad')
plt.show()
```

```python
# Gráfico de bastones (bar plot) para una variable categórica con valores numéricos
plt.figure(figsize=(12, 6))
sns.barplot(x='Nombre', y='Edad', data=df)
plt.title('Edades por Nombre')
plt.xlabel('Nombre')
plt.ylabel('Edad')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.countplot(x='Edad', data=df)
plt.title('Distribución de Edades')
plt.xlabel('Edaes')
plt.ylabel('Cantidad')
plt.show()
```

```python
# Gráfico de frecuencia para una variable categórica
plt.figure(figsize=(8, 6))
df['Ciudad'].value_counts().plot(kind='bar')
plt.title('Frecuencia de Ciudades')
plt.xlabel('Ciudad')
plt.ylabel('Frecuencia')
plt.show()
```

```python
# Diagrama de Tortas
tabla=df['Ciudad'].value_counts()
print(tabla)
plt.figure(figsize=(8, 6))
plt.pie(tabla,labels=tabla.index,autopct='%1.1f%%', startangle=140)
```
