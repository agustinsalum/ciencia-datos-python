```python
# Importando Librerias
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Funcion Lineal

x = np.linspace(0, 10, 100)
y = np.linspace(0,10,100)

# Crear el gráfico de líneas
plt.plot(x, y)
plt.title('Gráfico de Líneas')
plt.xlabel('Eje X')
plt.ylabel('Eje Y')
plt.grid(True)
plt.show()
```

```python
#Grafica de Barras
# Datos
categorias = ['A', 'B', 'C', 'D', 'E']
valores = [25, 18, 10, 32, 15]

# Crear el gráfico de barras
plt.bar(categorias, valores)
plt.title('Gráfico de Barras')
plt.xlabel('Categorías')
plt.ylabel('Valores')
plt.show()
```

```python
# Gráfico de Torta
categorias = ['A', 'B', 'C', 'D', 'E']
valores = [25, 18, 10, 32, 15]
plt.pie(valores, labels=categorias, autopct='%1.f%%')
plt.title('Gráfico de Torta')
plt.show()
```

```python
# Títulos y Etiquetas
x = np.linspace(0, 10, 100)
y = np.sin(x)
plt.plot(x, y)
plt.title('Gráfico de Líneas')
plt.xlabel('Eje X')
plt.ylabel('Eje Y')
plt.grid(True)
plt.show()
```

```python
# Colores y Estilos de Línea
x = np.linspace(-10, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
plt.plot(x, y1, 'r--', label='sin(x)')
plt.plot(x, y2, 'g-', label='cos(x)')
plt.legend()
plt.title('Gráfico de Líneas con Colores y Estilos')
plt.xlabel('Eje X')
plt.ylabel('Eje Y')
plt.grid(True)
plt.show()
```

```python
# Ejes y Rejillas
x = np.linspace(0, 20, 100)
y = np.sin(x)
plt.plot(x, y)
plt.title('Gráfico de Líneas')
plt.xlabel('Eje X')
plt.ylabel('Eje Y')
plt.xlim(0, 12)  # Límites del eje X
plt.ylim(-1.5, 1.5)  # Límites del eje Y
plt.grid(True)  # Mostrar rejilla
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Datos de ejemplo para el rango del eje X
x = np.linspace(-2, 2, 100)

# Función Lineal
y_lineal = x

# Función Cuadrática
y_cuadratica = x**2

# Función Logarítmica
y_logaritmica = np.log(x+6)

# Función Cúbica
y_cubica = x**3

# Crear el gráfico
plt.plot(x, y_lineal, label='Lineal')
plt.plot(x, y_cuadratica, label='Cuadrática')
plt.plot(x, y_logaritmica, label='Logarítmica')
plt.plot(x, y_cubica, label='Cúbica')

# Configuraciones del gráfico
plt.title('Funciones Lineal, Cuadrática, Logarítmica y Cúbica')
plt.xlabel('Eje X')
plt.ylabel('Eje Y')
plt.legend()  # Mostrar leyenda con las etiquetas de las funciones
plt.grid(True)

# Mostrar el gráfico
plt.show()
```

```python
# Datos de ejemplo para el rango del eje X
x = np.linspace(-5, 5, 100)

# Función Lineal
y_lineal = x

# Función Cuadrática
y_cuadratica = x**2

# Función Logarítmica
y_logaritmica = np.log(x + 6)

# Función Cúbica
y_cubica = x**3

# Crear el gráfico con subtramas
fig, axes = plt.subplots(2, 2, figsize=(10, 8))

# Subtrama 1: Función Lineal
axes[0, 0].plot(x, y_lineal, label='Lineal', color='blue')
axes[0, 0].set_title('Función Lineal')
axes[0, 0].set_xlabel('Eje X')
axes[0, 0].set_ylabel('Eje Y')
axes[0, 0].legend()
axes[0, 0].grid(True)

# Subtrama 2: Función Cuadrática
axes[0, 1].plot(x, y_cuadratica, label='Cuadrática', color='green')
axes[0, 1].set_title('Función Cuadrática')
axes[0, 1].set_xlabel('Eje X')
axes[0, 1].set_ylabel('Eje Y')
axes[0, 1].legend()
axes[0, 1].grid(True)

# Subtrama 3: Función Logarítmica
axes[1, 0].plot(x, y_logaritmica, label='Logarítmica', color='orange')
axes[1, 0].set_title('Función Logarítmica')
axes[1, 0].set_xlabel('Eje X')
axes[1, 0].set_ylabel('Eje Y')
axes[1, 0].legend()
axes[1, 0].grid(True)

# Subtrama 4: Función Cúbica
axes[1, 1].plot(x, y_cubica, label='Cúbica', color='red')
axes[1, 1].set_title('Función Cúbica')
axes[1, 1].set_xlabel('Eje X')
axes[1, 1].set_ylabel('Eje Y')
axes[1, 1].legend()
axes[1, 1].grid(True)

# Ajustar espacio entre subtramas
plt.tight_layout()

# Mostrar el gráfico con subtramas
plt.show()
```

```python
#Extras
# Gráfico 3D
from mpl_toolkits.mplot3d import Axes3D
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = np.sin(np.sqrt(X**2 + Y**2))
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z)
ax.set_title('Gráfico 3D')
ax.set_xlabel('Eje X')
ax.set_ylabel('Eje Y')
ax.set_zlabel('Eje Z')
plt.show()
```

```python
# Ejemplo de Gráfico Interactivo con Plotly

x = np.linspace(0, 10, 100)
y = np.sin(x)
fig = go.Figure()
fig.add_trace(go.Scatter(x=x, y=y, mode='lines', name='sin(x)'))
fig.update_layout(title='Gráfico de Líneas Interactivo con Plotly',
                  xaxis_title='Eje X',
                  yaxis_title='Eje Y')
fig.show()
```

```python
# Generar datos aleatorios para horas de estudio y puntaje en un examen
np.random.seed(42)  # Para reproducibilidad
horas_estudio = np.random.rand(50) * 10  # 50 valores aleatorios entre 0 y 10 horas
puntaje_examen = 50 + 4 * horas_estudio + np.random.normal(0, 5, 50)  # puntaje = 50 + 4 * horas_estudio + ruido gaussiano

# Crear el diagrama de dispersión
plt.scatter(horas_estudio, puntaje_examen, label='Datos')

# Agregar etiquetas de ejes y título
plt.xlabel('Horas de estudio')
plt.ylabel('Puntaje en el examen')
plt.title('Diagrama de dispersión: Horas de estudio vs Puntaje en el examen')




# Ajustar un modelo lineal
coef = np.polyfit(horas_estudio, puntaje_examen, 1)  # Ajuste de un polinomio de grado 1 (modelo lineal)
puntaje_predicho = coef[1] + coef[0] * horas_estudio  # Puntaje predicho utilizando el modelo

# Agregar la línea de regresión al gráfico
plt.plot(horas_estudio, puntaje_predicho, color='red', label='Modelo lineal')

# Mostrar el diagrama de dispersión y la línea de regresión
plt.legend()
plt.grid(True)
plt.show()

# Estimar el puntaje en función del número de horas de estudio
horas_estudio_desconocidas = np.array([5, 8, 10])
puntaje_estimado = coef[1] + coef[0] * horas_estudio_desconocidas
print("Puntaje estimado para las siguientes horas de estudio:")
for horas, puntaje in zip(horas_estudio_desconocidas, puntaje_estimado):
    print(f"Horas de estudio: {horas}, Puntaje estimado: {puntaje:.2f}")
```
