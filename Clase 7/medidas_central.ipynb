# **Medidas de tendencia Central**

Las medidas de tendencia central son herramientas estadísticas fundamentales utilizadas en la ciencia de datos para resumir y describir un conjunto de datos. Estas medidas proporcionan información sobre el valor típico o central de una distribución de datos. Las tres medidas de tendencia central más comunes son:

**Media (promedio):**

La media es la suma de todos los valores en el conjunto de datos dividida por el número de observaciones. Es una medida sensible a valores extremos (outliers) y proporciona una estimación general del centro de los datos.


```python
# Datos simulados de gastos mensuales de hogar de 10 personas
gastos_mensuales = [50000, 45000, 60000, 55000, 65000, 48000, 52000, 58000, 52000, 62000 ]

# Convertimos la lista de gastos_mensuales en un arreglo de numpy
gastos_mensuales_np = np.array(gastos_mensuales)

# Calculando la media (promedio) utilizando numpy
media = np.mean(gastos_mensuales_np)
```

```python
x = np.array(range(len(gastos_mensuales)))+1
plt.bar(x, gastos_mensuales)
plt.axhline(media, color='red', linestyle='dashed', linewidth=2, label='Media') # Dibuja la media en el eje y
plt.ylabel("Gastos Mensuales")
plt.xlabel("Personas")
plt.title("Gastos Mensuales de Personas Aleatorias")
plt.show()
```

La media de color rojo en forma horizontal. A continuacion, de forma vertical:

```python
# Graficar histograma para visualizar la distribución de los gastos mensuales
plt.figure(figsize=(8, 6))
plt.hist(gastos_mensuales_np, bins=5, edgecolor='black', alpha=0.7)
plt.axvline(media, color='red', linestyle='dashed', linewidth=2, label='Media (promedio)')
plt.xlabel('Gastos Mensuales de Hogar')
plt.ylabel('Frecuencia')
plt.title('Histograma de Gastos Mensuales de Hogar')
plt.legend()
plt.show()
```

Nota :

• La media es un estadístico sensible a valores extremos. Basta que algún dato
dentro de la muestra sea muy alto o muy bajo, el promedio se verá alterado.

**Mediana:**

La mediana es el valor que se encuentra en el centro de un conjunto de datos ordenado de menor a mayor. Es menos sensible a valores atípicos que la media y es útil cuando se tienen datos con distribuciones asimétricas.


Mediana: Se denomina Mediana ( Me )al número real tal que a lo sumo el 50% de los datos son
menores que él y a lo sumo el 50% son mayores.

```python
import numpy as np
import matplotlib.pyplot as plt

# Datos simulados de gastos mensuales de hogar de 10 personas
gastos_mensuales = [50000, 45000, 60000, 53000, 70000, 48000, 55000, 58000, 52000, 61000]

# Convertimos la lista de gastos_mensuales en un arreglo de numpy
gastos_mensuales_np = np.array(gastos_mensuales)

# Calculando la mediana utilizando numpy
mediana = np.median(gastos_mensuales_np)
```

```python
plt.scatter(gastos_mensuales,[1]*len(gastos_mensuales), color='blue')
# Dibuja en el eje y=1 , de color blue  y con el estilo '--'
plt.axhline(y=1, color='blue', linestyle='--')
plt.axvline(x=mediana, color='red', linestyle='--', label='Mediana')
plt.show()
```

Si un número par de elementos, se calcula tomando la media aritmética de los dos valores centrales después de ordenar el conjunto.

```python
# Graficar histograma para visualizar la distribución de los gastos mensuales
plt.figure(figsize=(8, 6))
plt.hist(gastos_mensuales_np, bins=5, edgecolor='black', alpha=0.7)
plt.axvline(mediana, color='red', linestyle='dashed', linewidth=2, label='Mediana')
plt.xlabel('Gastos Mensuales de Hogar')
plt.ylabel('Frecuencia')
plt.title('Histograma de Gastos Mensuales de Hogar')
plt.legend()
plt.show()
```

```python
# Generar datos aleatorios
np.random.seed(42)  # Fijamos una semilla para reproducibilidad
datos_aleatorios = np.random.normal(loc=50, scale=15, size=1000)

# Calcular la mediana
mediana = np.median(datos_aleatorios)

# Graficar histograma para visualizar la distribución de los datos
plt.figure(figsize=(8, 6))
plt.hist(datos_aleatorios, bins=30, edgecolor='black', alpha=0.7)
plt.axvline(mediana, color='red', linestyle='dashed', linewidth=2, label='Mediana')
plt.xlabel('Datos Aleatorios')
plt.ylabel('Frecuencia')
plt.title('Histograma de Datos Aleatorios')
plt.legend()
plt.show()

print("La mediana es:", mediana)
```

**Moda:**


La moda es el valor que aparece con mayor frecuencia en un conjunto de datos. Puede haber una sola moda (unimodal) o múltiples modas (multimodal). Es útil cuando se desea identificar el valor más frecuente en la distribución.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# Datos simulados de gastos mensuales de hogar de 10 personas
gastos_mensuales = [50000, 45000, 60000, 55000, 70000, 48000, 52000, 58000, 52000, 62000]

# Convertimos la lista de gastos_mensuales en un arreglo de numpy
gastos_mensuales_np = np.array(gastos_mensuales)

# Calculando la moda utilizando scipy.stats
moda = stats.mode(gastos_mensuales_np)[0][0]
# Graficar histograma para visualizar la distribución de los gastos mensuales
plt.figure(figsize=(8, 6))
plt.hist(gastos_mensuales_np, bins=5, edgecolor='black', alpha=0.7)
plt.axvline(moda, color='green', linestyle='dotted', linewidth=2, label='Moda')
plt.xlabel('Gastos Mensuales de Hogar')
plt.ylabel('Frecuencia')
plt.title('Histograma de Gastos Mensuales de Hogar')
plt.legend()
plt.show()
print("La moda es:", moda)
```

```python
# Crear una tabla de frecuencias
valores, frecuencias = np.unique(gastos_mensuales_np, return_counts=True)

# Graficar el diagrama de barras con barras más gruesas
plt.figure(figsize=(8, 6))
plt.bar(valores, frecuencias, edgecolor='black', alpha=0.7, width=10)
plt.axvline(moda, color='green', linestyle='dotted', linewidth=2, label='Moda')
plt.xlabel('Gastos Mensuales de Hogar')
plt.ylabel('Frecuencia')
plt.title('Diagrama de Barras de Gastos Mensuales de Hogar')
plt.legend()
plt.show()
print("La moda es:", moda)
```

# **Medidas de Dispersion**

Estas medidas se utilizan para cuantificar la variabilidad o extensión de un conjunto de datos. Indican cómo los valores individuales se distribuyen alrededor de una medida central, como la media o la mediana. Algunas medidas de dispersión comunes son la varianza, el desvío estándar, el rango, los percentiles y los cuartiles.

**Varianza**

La varianza es una medida de dispersión que representa la media de las diferencias al cuadrado entre cada valor individual en el conjunto de datos y la media del conjunto

np.var(conjunto_datos)

```python

import numpy as np
import matplotlib.pyplot as plt

# Datos con mucha varianza
conjuntosDatosA = [10, 15, 20, 5, 25, 30]

# Datos con poca varianza pero misma media
conjuntosDatosB = [20, 22, 18, 19, 21, 23]

# Calculando la media de cada conjunto de datos
mediaA = np.mean(conjuntosDatosA)
mediaB = np.mean(conjuntosDatosB)

# Calculando la varianza de cada conjunto de datos
varianza_a = np.var(conjuntosDatosA)
varianza_b = np.var(conjuntosDatosB)


print(f'Varianza para el Conjunto de Datos A {varianza_a} Para el conjuntos de datos B {varianza_b}')
```

```python
# Crear una lista de etiquetas para el eje Y con el mismo tamaño que los datos
etiquetas_mucha_varianza = [1] * len(conjuntosDatosA)
etiquetas_poca_varianza = [2] * len(conjuntosDatosB)

# Plot de los datos con mucha varianza en azul
plt.scatter(conjuntosDatosA, etiquetas_mucha_varianza, color='blue', label='Datos con mucha varianza')

# Plot de los datos con poca varianza en verde
plt.scatter(conjuntosDatosB, etiquetas_poca_varianza, color='green', label='Datos con poca varianza')

# Líneas horizontales que representan las medias de cada conjunto de datos
plt.axhline(y=1, color='blue', linestyle='--', label='Media datos con mucha varianza')
plt.axhline(y=2, color='green', linestyle='--', label='Media datos con poca varianza')
plt.legend()
plt.title('Comparación de Datos con Mucha Varianza y Poca Varianza')
plt.xlabel('Valor')
plt.yticks([1, 2], ['Datos con mucha varianza', 'Datos con poca varianza'])
plt.grid(True)
plt.show()
```

# Rango intercuartil

El rango intercuartil (RIC o IQR por sus siglas en inglés, Interquartile Range) es una medida estadística que proporciona información sobre la dispersión de los datos en la mitad central de un conjunto de datos. Es especialmente útil cuando se quiere describir la variabilidad de los datos sin verse afectado por valores extremos o atípicos que podrían estar presentes en el conjunto.

**Cuartiles**

Los cuartiles valores(Continuos/Discretos) que dividen la distribución de datos numéricos (ordenados de
menor a mayor) en cuatro partes (los cuartos) que corresponden al 25% cada una:

<img src="imagenes/cuartiles.png">

El primer cuartil (Q1) representa el valor que separa el 25% de los datos más bajos del 75% de los datos más altos.
La mediana (Q2) representa el valor que separa el 50% de los datos más bajos del 50% de los datos más altos. En otras palabras, es el valor en el centro del conjunto de datos ordenados.
El tercer cuartil (Q3) representa el valor que separa el 75% de los datos más bajos del 25% de los datos más altos.

```python
# Datos de ejemplo
estaturas = np.array([150, 170, 175, 168, 180, 165, 172, 178, 182, 169, 176, 171, 174, 167, 179, 173, 177, 166, 181, 183, 168, 200])

# Calcula los cuartiles
q1 = np.percentile(estaturas, 25)
q2 = np.percentile(estaturas, 50)  # Mediana
q3 = np.percentile(estaturas, 75)

# Gráfica de caja y bigotes (boxplot)
plt.xticks([1], ['Datos'])
plt.grid(True)

# Marca los cuartiles en la gráfica
plt.scatter(1, q1, color='red', label='Primer cuartil (Q1)')
plt.scatter(1, q2, color='green', label='Mediana (Q2)')
plt.scatter(1, q3, color='blue', label='Tercer cuartil (Q3)')

plt.legend()
plt.show()
```

La diferencia entre el tercer cuartil y el primer cuartil se llama rango entre cuartiles, denotado por RQ=Q3-Q1. El rango entre cuartiles mide la variabilidad de la mitad central de los datos

```python
# Calcula el rango intercuartil (IQR)
iqr = q3 - q1

print("Rango intercuartil (IQR):", iqr)
```

* Rango intercuartil alto:
Si el RIC es alto, significa que los datos en la mitad central del conjunto están más dispersos y se extienden por un rango amplio de valores. En otras palabras, existe una gran variabilidad entre los valores que se encuentran en el segundo y tercer cuartil (Q2 y Q3). Esto puede indicar que hay una gran diversidad en las observaciones dentro de esta parte central del conjunto de datos.
* Rango intercuatil Bajo:
Si el RIC es bajo, indica que los datos en la mitad central del conjunto están relativamente cerca unos de otros, lo que sugiere una menor dispersión y una mayor concentración de valores alrededor de la mediana. Esto puede implicar una mayor consistencia o uniformidad en los datos dentro de esta parte del conjunto.


# El Diagrama de Caja

El Diagrama de Caja es una representación gráfica que utiliza el rango intercuartil

Este tipo de gráficos nos permite identificar valores atípicos y comparar distribuciones. Además de conocer de una forma cómoda y rápida como el 50% de los valores centrales se distribuyen.

El diagrama de cajas muestra un resumen estadístico de los datos mediante una representación gráfica que incluye cinco medidas principales: el mínimo, el primer cuartil (Q1), la mediana (Q2), el tercer cuartil (Q3) y el máximo

Componentes
* Rectángulo (caja):

* Bigotes (líneas verticales):

* Valores atípicos (outliers):

<img src="imagenes/cuartiles_2.png">

```python

# Vector con estaturas específicas
estaturas = np.array([150, 170, 175, 168, 180, 165, 172, 178, 182, 169, 176, 171, 174, 167, 179, 173, 177, 166, 181, 183, 168,200])

# Crear el boxplot de las estaturas
plt.figure(figsize=(10, 5))  # Definimos el tamaño de la figura
plt.boxplot(estaturas, vert=True, labels=["Estaturas"])  # Vert=True para orientar verticalmente el boxplot

# Estilo y etiquetas
plt.ylabel("Estatura (cm)", fontsize=14)  # Agregamos el ylabel con un tamaño de fuente de 14
plt.yticks(np.arange(150, 201, 5))  # Etiquetas personalizadas para el eje y con intervalos de 5
plt.ylim(145, 205)  # Ajustamos los límites del eje y para asegurar que el boxplot se visualice correctamente
plt.title("Boxplot de Estaturas", fontsize=16)  # Título con un tamaño de fuente de 16
plt.grid(True)  # Agregamos una rejilla solo en el eje x para mayor claridad

# Mostramos el gráfico
plt.show()
```

```python
estaturas = np.array([150, 170, 175, 168, 180, 165, 172, 178, 182, 169, 176, 171, 174, 167, 179, 173, 177, 166, 181, 183, 168, 200])

# Crear el boxplot de las estaturas
plt.figure(figsize=(10, 5))  # Definimos el tamaño de la figura
plt.boxplot(estaturas, vert=True, labels=["Estaturas"])  # Vert=True para orientar verticalmente el boxplot

# Mostrar todos los datos como puntos (outliers)
plt.plot(np.random.normal(1, 0.04, size=len(estaturas)), estaturas, 'ro', alpha=0.6)

# Estilo y etiquetas
plt.ylabel("Estatura (cm)", fontsize=14)  # Agregamos el ylabel con un tamaño de fuente de 14
plt.yticks(np.arange(150, 201, 5))  # Etiquetas personalizadas para el eje y con intervalos de 5
plt.ylim(145, 205)  # Ajustamos los límites del eje y para asegurar que el boxplot se visualice correctamente
plt.title("Boxplot de Estaturas", fontsize=16)  # Título con un tamaño de fuente de 16
plt.grid(True)  # Agregamos una rejilla solo en el eje x pa mayor claridad

# Mostramos el gráfico
plt.show()
```

Interpretación general:

* Si la caja es corta y los bigotes son cortos, indica que los datos están concentrados y la dispersión es baja.
* Si la caja es larga y los bigotes son largos, indica que los datos están más dispersos.
* Si los bigotes son asimétricos o tienen diferentes longitudes, puede indicar una distribución sesgada.
* Si hay valores atípicos, es importante investigarlos y comprender su impacto en el análisis.
