## Que tipo de variables son. Justificar tu respuesta

Las variables se pueden clasificar en cuatro categorías principales: nominales, ordinales, discretas y continuas. Aquí hay una explicación breve de cada una:

- Variables Nominales: Estas variables representan categorías o grupos que no tienen un orden inherente. No hay relación de tamaño o magnitud entre las categorías. Ejemplos incluyen el género, el estado civil, el color favorito, etc.

- Variables Ordinales: Estas variables también representan categorías, pero tienen un orden específico entre ellas. Sin embargo, la diferencia en la magnitud entre las categorías no es significativa o constante. Ejemplos incluyen la calificación de satisfacción (baja, media, alta), nivel de educación (primaria, secundaria, universitaria), etc.

- Variables Discretas: Estas variables toman valores separados y cuentan cantidades específicas. Los valores son generalmente números enteros y no pueden tener valores intermedios. Ejemplos incluyen la cantidad de hijos, el número de autos en un hogar, el número de estudiantes en una clase, etc.

- Variables Continuas: Estas variables pueden tomar cualquier valor dentro de un rango determinado y tienen infinitos valores posibles. En otras palabras, pueden tener valores decimales y se miden en una escala continua. Ejemplos incluyen la altura, el peso, la temperatura, el ingreso, etc.

Aquí hay algunos ejemplos para ilustrar mejor estas categorías:

- Nominal: Género (hombre, mujer), Color de ojos (azul, verde, marrón), País de origen (EE. UU., Canadá, México).

- Ordinal: Calificación de satisfacción (baja, media, alta), Nivel de educación (primaria, secundaria, universitaria), Clasificación de estrellas (1 estrella, 2 estrellas, 3 estrellas, etc.).

- Discreta: Número de hijos en una familia, Cantidad de libros en una biblioteca, Número de estudiantes en una clase.

- Continua: Altura de una persona, Peso corporal, Ingreso anual, Temperatura en grados Celsius.

### Valores Ausentes:

Los valores ausentes son datos que faltan en una o más observaciones de un conjunto de datos. Estos valores pueden haberse perdido por diversas razones, como errores de entrada, problemas de medición o simplemente porque no se registraron. La presencia de valores ausentes puede afectar negativamente el análisis de datos y la precisión de los modelos.

- Identificacion: Valores ausentes en un DataFrame de pandas, puedes usar el método .isna() o .isnull(). Estos métodos devuelven una matriz booleana del mismo tamaño que el DataFrame, donde True representa un valor ausente y False representa un valor presente. Luego, puedes utilizar funciones agregadas como .sum() para contar la cantidad de valores ausentes en cada columna. 

- Estrategias:
    - Eliminación de Filas: Si los valores ausentes son pocos y no afectan significativamente tu conjunto de datos, puedes optar por eliminar las filas que contengan estos valores.

    - Reemplazar por cero: En caso de realizar una sumatoria puede optar por reemplazar por cero los datos faltantes

    - Imputación: Rellenar los valores ausentes con estimaciones razonables. Esto puede hacerse utilizando la media, la mediana o una técnica más avanzada, como la imputación basada en modelos de regresión.

    - Usar marcadores: En lugar de imputar valores reales, puedes asignar valores marcadores a los valores ausentes para indicar que están faltantes.

```python
# Obtener el número de valores faltantes en cada columna
print ("Cantida de valores faltantes de Diabetes: ")
df_diabetes.isnull().sum()
```

```python
print ("Cantida de valores faltantes de Iris: ")
df_iris.isnull().sum()
```

```python
print ("Cantida de valores faltantes de Titanic: ")
df_titanic.isnull().sum()
```

```python
print ("Cantida de valores faltantes de Wine: ")
df_wineqt.isnull().sum()
```

Como se puede apreciar, existen datos faltantes en las variables edad, cabina y embarcacion del Titanic. El tratamiento mas adecuado sera reemplazar las edades por la mediana y reemplazar por la moda las variables faltantes en "cambina" y "embarcacion".
En caso de que se trate de una variable la cual debe aplicarse la sumatoria, la mejor opcion es reemplazar los datos faltantes por cero.

```python
# Rellenar los valores faltantes en la columna 'Age' con la mediana
df_titanic['Age'].fillna(df_titanic['Age'].median(), inplace=True)

# Rellenar los valores faltantes en la columna 'Embarked' con el valor más frecuente
frecuente_embarked = df_titanic['Embarked'].mode()[0]
frecuente_cabin = df_titanic['Cabin'].mode()[0]

df_titanic['Embarked'].fillna(frecuente_embarked, inplace=True)
df_titanic['Cabin'].fillna(frecuente_cabin, inplace=True)

# Verificamos
df_titanic.isnull().sum()

```

### Valores Atípicos:

Los valores atípicos son observaciones que difieren significativamente del resto de los datos en un conjunto. Estos valores representan situaciones excepcionales en los datos. Los valores atípicos pueden influir en los resultados del análisis y en la interpretación de los modelos.

- Identificación: Utiliza métodos estadísticos o gráficos (como diagramas de caja) para identificar los valores atípicos en tus datos.

- Tratamientos: Decide si deseas eliminar los valores atípicos, ajustarlos a un valor razonable o dejarlos como están, dependiendo de la naturaleza de tus datos y los objetivos de tu análisis.

    - Truncamiento: Cambia los valores atípicos a un valor máximo o mínimo aceptable para evitar que afecten de manera negativa los resultados.
    
    - Transformación de datos: Aplica transformaciones a los datos para reducir la influencia de los valores atípicos. Por ejemplo, una transformación logarítmica puede ayudar a suavizar la distribución.
    
    - Eliminación: En algunos casos, cuando los valores atípicos son claramente errores o anomalías, puedes optar por eliminarlos. Sin embargo, debes ser cuidadoso con esta estrategia, ya que los valores atípicos legítimos también pueden contener información valiosa.

Generalmente se recomienda crear diagramas de cajas (box plots) para cada variable que quieras analizar en busca de valores atípicos. Sin embargo, si tienes un gran número de variables y no todas son igualmente relevantes para tu análisis, podrías considerar enfocarte en las variables más críticas o aquellas que consideres más propensas a tener valores atípicos.
Los diagramas de cajas son solo una herramienta visual para identificar valores atípicos y que también puedes combinarlos con métodos estadísticos para obtener una comprensión más completa de la presencia y el impacto de los valores atípicos en tus datos.







```python
import seaborn as sns

# Configurar el estilo de los gráficos con Seaborn
sns.set(style="whitegrid")

# Crear un diagrama de cajas para la columna 'Age'
plt.figure(figsize=(8, 6))
sns.boxplot(data=df_titanic, x='Age')
plt.title('Diagrama de Cajas para la Edad')
plt.xlabel('Edad')
plt.show()
```

Se observan muchos puntos fuera del rango intercuartil (IQR) para la variable "Age", es probable que sean valores atípicos. En este caso no tiene sentido aplicar un tratamiento para las edades atipicas encontradas

#### Gráfico de Línea:
- Variables adecuadas: Variables continuas en el tiempo, como series temporales.
- Justificación: Este tipo de gráfico es útil para mostrar tendencias y cambios a lo largo del tiempo.

#### Histograma:
- Variables adecuadas: Variables continuas o discretas.
- Justificación: Los histogramas son ideales para visualizar la distribución de los datos y ayudan a entender la frecuencia de diferentes intervalos.

#### Gráfico de Barras:
- Variables adecuadas: Variables nominales u ordinales.
- Justificación: Los gráficos de barras permiten comparar diferentes categorías y mostrar la frecuencia o el valor asociado con cada una de ellas.

#### Gráfico de Pastel o Torta:
- Variables adecuadas: Variables nominales.
- Justificación: Los gráficos de pastel son útiles para mostrar proporciones relativas dentro de un conjunto de categorías.

#### Box-Plot:
- Variables adecuadas: Variables continuas.
- Justificación: Los box-plots son excelentes para visualizar la distribución, mediana, cuartiles y valores atípicos de una variable continua.

- Variables adecuadas: Variables continuas.
- Justificación: Los gráficos de violín combinan un box-plot con una representación de la densidad, lo que ayuda a mostrar la distribución de los datos de manera más detallada.

#### Gráfico de Dispersión o Scatter Plot:
- Variables adecuadas: Pares de variables continuas.
- Justificación: Los scatter plots son ideales para visualizar la relación entre dos variables continuas y detectar patrones como correlaciones.

#### Mapa de Calor:
- Variables adecuadas: Matrices de datos, como correlaciones entre variables.
- Justificación: Los mapas de calor resaltan relaciones y patrones en datos matriciales, siendo útiles para mostrar la intensidad de una medida entre diferentes pares de variables.

- Variables adecuadas: Variables categóricas.
- Justificación: Los diagramas de árbol son útiles para representar estructuras jerárquicas o de decisión en datos categorizados.

- Variables adecuadas: Variables relacionales, como conexiones entre entidades.
- Justificación: Los gráficos de red son útiles para visualizar interacciones y relaciones entre diferentes elementos o nodos.

El box-plot nos indica que existen muchos valores atípicos (outliers). Estos son observaciones que se desvían significativamente del resto de los datos en un conjunto. Su desventaja principal es que pueden afectar negativamente el análisis y la interpretación de los datos.

```python
import pandas as pd

# Primer Cuartil (Q1) y el Tercer Cuartil (Q3)
Q1 = df_titanic["Age"].quantile(0.25) # valor por debajo del cual se encuentra el 25% de los datos
Q3 = df_titanic["Age"].quantile(0.75) # valor por debajo del cual se encuentra el 75% de los datos
# Rango Intercuartil (IQR)
IQR = Q3 - Q1

# Límites para Detectar Valores Atípicos
limite_inferior = Q1 - 1.5 * IQR
limite_superior = Q3 + 1.5 * IQR

# Los valores que caen por debajo del límite inferior o por encima del límite superior son valores atípicos
valores_atipicos = df_titanic[(df_titanic["Age"] < limite_inferior) | (df_titanic["Age"] > limite_superior)]

valores_atipicos
```

## Realizar analisis de corelacion y explicar que variable estan correlacionadas



```python
# Correlación entre variables
correlation_matrix = df_diabetes.corr()
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Matriz de correlación')
plt.show()
```

los coeficientes de correlación indican la fuerza y dirección de la relación lineal entre las variables. Un valor cercano a 1 o -1 indica una correlación fuerte (positiva o negativa, respectivamente), mientras que un valor cercano a 0 indica una correlación débil o la ausencia de una relación lineal.
Cuando dos variables tienen una correlación fuerte positiva, significa que tienden a aumentar juntas. En otras palabras, cuando el valor de una variable aumenta, es probable que el valor de la otra variable también aumente.
En el caso de una correlación fuerte negativa, cuando el valor de una variable aumenta, el valor de la otra variable tiende a disminuir.

    - Es relativamente bajo. Existe una relación positiva débil entre estas dos variables. Esto significa que a medida que el número de embarazos aumenta, tiende a haber un aumento leve en la concentración de glucosa en plasma.
        
- El coeficiente de correlación entre "BMI" y sí mismo es 1.00
    - Es el máximo posible y significa que cada variable se correlaciona perfectamente consigo misma. Se considera redundante.

- El coeficiente de correlación entre "Pregnancies" y "DiabetesPedigreeFunction" es de: -0.03
    - Un coeficiente de correlación cercano a cero indica que no hay una relación lineal evidente entre estas dos variables. La relación negativa muy débil sugiere que no hay una relación clara en la que el número de embarazos influya en la función de predisposición genética a la diabetes.

- El coficinetre de correlacion entre Outcome con Glucose es de: 0.47
    - Un coeficiente de correlación de 0.47 indica una relación positiva moderada entre estas dos variables.
    La concentración de glucosa en plasma ("Glucose") tiene una correlación moderada con la variable objetivo ("Outcome"), que indica si la paciente tiene diabetes o no. El valor sugiere que a medida que la concentración de glucosa en plasma aumenta, hay una tendencia a que la paciente tenga un diagnóstico positivo de diabetes (Outcome = 1).
