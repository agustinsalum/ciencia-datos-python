## Que tipo de variables son. Justificar tu respuesta

Las variables se pueden clasificar en cuatro categorÃ­as principales: nominales, ordinales, discretas y continuas. AquÃ­ hay una explicaciÃ³n breve de cada una:

- Variables Nominales: Estas variables representan categorÃ­as o grupos que no tienen un orden inherente. No hay relaciÃ³n de tamaÃ±o o magnitud entre las categorÃ­as. Ejemplos incluyen el gÃ©nero, el estado civil, el color favorito, etc.

- Variables Ordinales: Estas variables tambiÃ©n representan categorÃ­as, pero tienen un orden especÃ­fico entre ellas. Sin embargo, la diferencia en la magnitud entre las categorÃ­as no es significativa o constante. Ejemplos incluyen la calificaciÃ³n de satisfacciÃ³n (baja, media, alta), nivel de educaciÃ³n (primaria, secundaria, universitaria), etc.

- Variables Discretas: Estas variables toman valores separados y cuentan cantidades especÃ­ficas. Los valores son generalmente nÃºmeros enteros y no pueden tener valores intermedios. Ejemplos incluyen la cantidad de hijos, el nÃºmero de autos en un hogar, el nÃºmero de estudiantes en una clase, etc.

- Variables Continuas: Estas variables pueden tomar cualquier valor dentro de un rango determinado y tienen infinitos valores posibles. En otras palabras, pueden tener valores decimales y se miden en una escala continua. Ejemplos incluyen la altura, el peso, la temperatura, el ingreso, etc.

AquÃ­ hay algunos ejemplos para ilustrar mejor estas categorÃ­as:

- Nominal: GÃ©nero (hombre, mujer), Color de ojos (azul, verde, marrÃ³n), PaÃ­s de origen (EE. UU., CanadÃ¡, MÃ©xico).

- Ordinal: CalificaciÃ³n de satisfacciÃ³n (baja, media, alta), Nivel de educaciÃ³n (primaria, secundaria, universitaria), ClasificaciÃ³n de estrellas (1 estrella, 2 estrellas, 3 estrellas, etc.).

- Discreta: NÃºmero de hijos en una familia, Cantidad de libros en una biblioteca, NÃºmero de estudiantes en una clase.

- Continua: Altura de una persona, Peso corporal, Ingreso anual, Temperatura en grados Celsius.

### Valores Ausentes:

Los valores ausentes son datos que faltan en una o mÃ¡s observaciones de un conjunto de datos. Estos valores pueden haberse perdido por diversas razones, como errores de entrada, problemas de mediciÃ³n o simplemente porque no se registraron. La presencia de valores ausentes puede afectar negativamente el anÃ¡lisis de datos y la precisiÃ³n de los modelos.

- Identificacion: Valores ausentes en un DataFrame de pandas, puedes usar el mÃ©todo .isna() o .isnull(). Estos mÃ©todos devuelven una matriz booleana del mismo tamaÃ±o que el DataFrame, donde True representa un valor ausente y False representa un valor presente. Luego, puedes utilizar funciones agregadas como .sum() para contar la cantidad de valores ausentes en cada columna. 

- Estrategias:
    - EliminaciÃ³n de Filas: Si los valores ausentes son pocos y no afectan significativamente tu conjunto de datos, puedes optar por eliminar las filas que contengan estos valores.

    - Reemplazar por cero: En caso de realizar una sumatoria puede optar por reemplazar por cero los datos faltantes

    - ImputaciÃ³n: Rellenar los valores ausentes con estimaciones razonables. Esto puede hacerse utilizando la media, la mediana o una tÃ©cnica mÃ¡s avanzada, como la imputaciÃ³n basada en modelos de regresiÃ³n.

    - Usar marcadores: En lugar de imputar valores reales, puedes asignar valores marcadores a los valores ausentes para indicar que estÃ¡n faltantes.

```python
# Obtener el nÃºmero de valores faltantes en cada columna
print ("Cantida de valores faltantes de Diabetes: ")
df_diabetes.isnull().sum()
```

```python
print ("Cantida de valores faltantes de Iris: ")
df_iris.isnull().sum()
```

```python
print ("Cantida de valores faltantes de Titanic: ")
df_titanic.isnull().sum()
```

```python
print ("Cantida de valores faltantes de Wine: ")
df_wineqt.isnull().sum()
```

Como se puede apreciar, existen datos faltantes en las variables edad, cabina y embarcacion del Titanic. El tratamiento mas adecuado sera reemplazar las edades por la mediana y reemplazar por la moda las variables faltantes en "cambina" y "embarcacion".
En caso de que se trate de una variable la cual debe aplicarse la sumatoria, la mejor opcion es reemplazar los datos faltantes por cero.

```python
# Rellenar los valores faltantes en la columna 'Age' con la mediana
df_titanic['Age'].fillna(df_titanic['Age'].median(), inplace=True)

# Rellenar los valores faltantes en la columna 'Embarked' con el valor mÃ¡s frecuente
frecuente_embarked = df_titanic['Embarked'].mode()[0]
frecuente_cabin = df_titanic['Cabin'].mode()[0]

df_titanic['Embarked'].fillna(frecuente_embarked, inplace=True)
df_titanic['Cabin'].fillna(frecuente_cabin, inplace=True)

# Verificamos
df_titanic.isnull().sum()

```

### Valores AtÃ­picos:

Los valores atÃ­picos son observaciones que difieren significativamente del resto de los datos en un conjunto. Estos valores representan situaciones excepcionales en los datos. Los valores atÃ­picos pueden influir en los resultados del anÃ¡lisis y en la interpretaciÃ³n de los modelos.

- IdentificaciÃ³n: Utiliza mÃ©todos estadÃ­sticos o grÃ¡ficos (como diagramas de caja) para identificar los valores atÃ­picos en tus datos.

- Tratamientos: Decide si deseas eliminar los valores atÃ­picos, ajustarlos a un valor razonable o dejarlos como estÃ¡n, dependiendo de la naturaleza de tus datos y los objetivos de tu anÃ¡lisis.

    - Truncamiento: Cambia los valores atÃ­picos a un valor mÃ¡ximo o mÃ­nimo aceptable para evitar que afecten de manera negativa los resultados.
    
    - TransformaciÃ³n de datos: Aplica transformaciones a los datos para reducir la influencia de los valores atÃ­picos. Por ejemplo, una transformaciÃ³n logarÃ­tmica puede ayudar a suavizar la distribuciÃ³n.
    
    - EliminaciÃ³n: En algunos casos, cuando los valores atÃ­picos son claramente errores o anomalÃ­as, puedes optar por eliminarlos. Sin embargo, debes ser cuidadoso con esta estrategia, ya que los valores atÃ­picos legÃ­timos tambiÃ©n pueden contener informaciÃ³n valiosa.

Generalmente se recomienda crear diagramas de cajas (box plots) para cada variable que quieras analizar en busca de valores atÃ­picos. Sin embargo, si tienes un gran nÃºmero de variables y no todas son igualmente relevantes para tu anÃ¡lisis, podrÃ­as considerar enfocarte en las variables mÃ¡s crÃ­ticas o aquellas que consideres mÃ¡s propensas a tener valores atÃ­picos.
Los diagramas de cajas son solo una herramienta visual para identificar valores atÃ­picos y que tambiÃ©n puedes combinarlos con mÃ©todos estadÃ­sticos para obtener una comprensiÃ³n mÃ¡s completa de la presencia y el impacto de los valores atÃ­picos en tus datos.







```python
import seaborn as sns

# Configurar el estilo de los grÃ¡ficos con Seaborn
sns.set(style="whitegrid")

# Crear un diagrama de cajas para la columna 'Age'
plt.figure(figsize=(8, 6))
sns.boxplot(data=df_titanic, x='Age')
plt.title('Diagrama de Cajas para la Edad')
plt.xlabel('Edad')
plt.show()
```

Se observan muchos puntos fuera del rango intercuartil (IQR) para la variable "Age", es probable que sean valores atÃ­picos. En este caso no tiene sentido aplicar un tratamiento para las edades atipicas encontradas

El box-plot nos indica que existen muchos valores atÃ­picos (outliers). Estos son observaciones que se desvÃ­an significativamente del resto de los datos en un conjunto. Su desventaja principal es que pueden afectar negativamente el anÃ¡lisis y la interpretaciÃ³n de los datos.

```python
import pandas as pd

# Primer Cuartil (Q1) y el Tercer Cuartil (Q3)
Q1 = df_titanic["Age"].quantile(0.25) # valor por debajo del cual se encuentra el 25% de los datos
Q3 = df_titanic["Age"].quantile(0.75) # valor por debajo del cual se encuentra el 75% de los datos
# Rango Intercuartil (IQR)
IQR = Q3 - Q1

# LÃ­mites para Detectar Valores AtÃ­picos
limite_inferior = Q1 - 1.5 * IQR
limite_superior = Q3 + 1.5 * IQR

# Los valores que caen por debajo del lÃ­mite inferior o por encima del lÃ­mite superior son valores atÃ­picos
valores_atipicos = df_titanic[(df_titanic["Age"] < limite_inferior) | (df_titanic["Age"] > limite_superior)]

valores_atipicos
```
