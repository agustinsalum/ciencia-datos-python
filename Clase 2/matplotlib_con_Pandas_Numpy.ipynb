```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

Primera gráfica lineal con Matplitlib: Gráfica sencilla X-Y de dos arrays

```python
# Una línea de código en Python que usa la función numpy.linspace ()
# para crear una matriz de 100 números espaciados uniformemente entre 0 y 10.

x = np.linspace(0, 10, 100)
x
```

La función numpy.linspace() toma tres argumentos:

- el valor inicial
- el valor final
- el número de muestras que se desea generar.
- El valor inicial es 0, el valor final es 10 y el número de muestras es 100.

La función devuelve una matriz de números que están distribuidos uniformemente sobre el intervalo [0, 10].

Podemos usar esta función para crear una secuencia de números con un espaciado constante. Por ejemplo, podemos usarla para generar puntos para trazar una función matemática.

```python
# Mostramos un arreglo de números que representan la función seno de otro arreglo.

y = np.sin(x)
y

# El arreglo x se crea con la función np.linspace(),
# que genera una secuencia de números espaciados uniformemente entre dos límites.
# En este caso, el arreglo x tiene 100 números entre 0 y 10.
# Luego, el arreglo y se crea con la función numpy.sin(),
# que calcula el seno de cada elemento del arreglo x.
```

# Qué es la función seno

El seno es una función trigonométrica que relaciona el ángulo de un triángulo rectángulo con la razón entre el cateto opuesto y la hipotenusa.

Finalmente, el arreglo y se muestra en la pantalla con la función print().

El resultado es un arreglo de números entre -1 y 1 que corresponden al seno de los números del arreglo x.

```python
plt.plot(x, y)
plt.show()
```

# Gráfica lineal "completa"

Cómo darle nombre a los ejes, título de la gráfica, añadir rejilla, modificar el tamaño...

```python
# Tamaño
plt.figure(figsize=(15, 7))

# Trazar
plt.plot(x, y)

# Titulo
plt.title('f(x) -> seno(x)', fontsize=20)

# Tipografia de 20
plt.xlabel('x', fontsize=20)
plt.ylabel('seno(x)', fontsize=20)

# Establecer rango
plt.xlim(0, 10)
plt.ylim(-1.1, 1.1)

# Establecemos el tamaño de la fuente de las marcas del eje y a 15 puntos.

plt.yticks(fontsize=15)

# Establecemos el tamaño de la fuente de las marcas del eje x a 15 puntos
# y las posiciones de las marcas a intervalos de 0.5 entre 0 y 10.

plt.xticks(np.arange(0, 10.5, 0.5), fontsize=15)

# Añadimos una rejilla al gráfico para facilitar la lectura de los valores.

plt.grid()

# Guardamos el gráfico en un archivo llamado funcion_seno.png en el directorio actual.

plt.savefig('funcion_seno.png')

plt.show()
```

# Subplots

Varias gráficas en un mismo lienzo, pero diferentes ejes

```python
# Creamos un panel de gráficos con dos filas y una columna,
# y seleccionamos el primer panel (el de arriba) para dibujar el primer gráfico.

plt.subplot(2, 1, 1) # (filas, columnas, número de panel)
plt.plot(x, y) # trazamos la función seno

# Seleccionamos el segundo panel (el de abajo) para dibujar el segundo gráfico.
plt.subplot(2, 1, 2)
plt.plot(x, y, color='r')

plt.show()
```

```python
# Trazamos el gráfico de la función seno de x,
# usando los arreglos x e y creados anteriormente con las funciones numpy.linspace() y numpy.sin().

plt.plot(x, y)

# Creamos un nuevo arreglo y2 con los valores de la función coseno de x,
# usando la función numpy.cos() que calcula el coseno de cada elemento del arreglo x.

y2 = np.cos(x)

# Trazamos el gráfico de la función coseno de x, usando los arreglos x e y2,
# y especificando el estilo de línea como ‘–’, que significa una línea discontinua.

plt.plot(x, y2, '--')

plt.show()
```

# Subplots

Con títulos, etiqueta de ejes...

```python
# Creamos una figura con dos subgráficos dispuestos en una fila y dos columnas,
# y asignamos los objetos ax1 y ax2 a cada subgráfico.
# El argumento figsize especifica el tamaño de la figura en pulgadas.

fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2,figsize=(14, 5))

# Dibujamos un diagrama de dispersión usando los arreglos x e y
# creados anteriormente con las funciones numpy.linspace() y numpy.sin(),
# y asignamos el color rojo a los puntos, el color azul a los bordes
# y la forma de círculo al marcador.

ax1.scatter(x=x, y=y, marker='o', c='r', edgecolor='b')

# Establecemos el título del primer subgráfico como ‘Scatter: x versus sen(x)’.

ax1.set_title('Scatter: x versus sen(x)')

# Establecemos la etiqueta del eje x del primer subgráfico como ‘x’.
# El símbolo ‘$’ indica que se usa la notación matemática LaTeX.

ax1.set_xlabel('$x$')
ax1.set_ylabel('$seno(x)$')

# Dibujamos un gráfico de línea usando los arreglos x e y2 creados anteriormente
# con la función numpy.cos(),
# y asignamos el color rojo a la línea y la etiqueta ‘cos(x)’ a la leyenda.

ax2.plot(x, y2, color='r', label=('cos(x)'))

# Muestra la leyenda del segundo subgráfico en la posición óptima.

ax2.legend(loc='best')

# Establecemos el título del segundo subgráfico como ‘Gráfico lineal x versus cos(x)’.

ax2.set_title('Gráfico lineal x versus cos(x)')

# Movemos las marcas del eje y del segundo subgráfico al lado derecho.

ax2.yaxis.tick_right()

# Establecemos la etiqueta del eje x del segundo subgráfico como ‘x’.
# El símbolo ‘$’ indica que se usa la notación matemática LaTeX.

ax2.set_xlabel('$x$')

# Establecemos la etiqueta del eje y del segundo subgráfico como ‘coseno(x)’.
# El símbolo ‘$’ indica que se usa la notación matemática LaTeX.

ax2.set_ylabel('$coseno(x)$')

plt.show()
```

# Histograma

# Gráfico de barras



```python
# agrupa el dataframe df por la columna ‘marital-status’ y cuenta el número de personas por cada categoría. El resultado es un nuevo dataframe con dos columnas: ‘marital-status’ y ‘age’. El método reset_index() elimina el índice jerárquico creado por el método groupby() y lo reemplaza por un índice simple.

df_group = df.groupby(['marital-status'])['age'].count().reset_index()

# cambia el nombre de la columna ‘age’ por ‘number’, para que sea más claro que representa el número de personas por estado civil.

df_group.columns = ['marital-status', 'number']

# ordena el dataframe df_group por la columna ‘number’ de forma descendente, es decir, de mayor a menor. Así se puede ver fácilmente cuál es el estado civil más frecuente y cuál es el menos frecuente.

df_group = df_group.sort_values(by=['number'], ascending=False)
```

```python
# crea un gráfico de barras usando los valores de las columnas ‘marital-status’ y ‘number’ del dataframe df_group, y asigna el color naranja a las barras.

plt.bar(df_group['marital-status'], df_group['number'], color='orange')

# rota las etiquetas del eje x 70 grados, para que se puedan leer mejor las categorías del estado civil.

plt.xticks(rotation=70)

# establece la etiqueta del eje x como ‘Estado civil’.
plt.xlabel('Estado civil')

plt.ylabel('Nº. de casos')

plt.title('Nº de personas por estado civil')
plt.show()
```

# Gráfico de barras horizontal

```python
# crea un gráfico de barras horizontal usando los valores de las columnas ‘marital-status’
# y ‘number’ del dataframe df_group, y asigna el color naranja a las barras con una
# transparencia de 0.3. El argumento alpha especifica el nivel de opacidad de las barras,
# donde 0 es totalmente transparente y 1 es totalmente opaco.

plt.barh(df_group['marital-status'], df_group['number'], color='orange', alpha=0.3)

# rota las etiquetas del eje x 70 grados, para que se puedan leer mejor los valores numéricos.

plt.xticks(rotation=70)

plt.xlabel('Nº. de casos')
plt.ylabel('Estado civil')
plt.title('Nº de personas por estado civil')

plt.show()
```

# Barras con diferentes propiedades

```python
df_group = df.groupby(['marital-status', 'salary'])['age'].count().reset_index()
df_group.columns = ['marital-status', 'salary', 'number']
df_group = df_group.sort_values(by=['number'], ascending=False)
```

```python
df_group_menor = df_group[df_group.salary==" <=50K"]
df_group_mayor = df_group[df_group.salary==" >50K"]
plt.bar(df_group_menor['marital-status'], df_group_menor['number'], color='orange', label=' <=50K')
plt.bar(df_group_mayor['marital-status'], df_group_mayor['number'],label=' >50K')
plt.xticks(rotation=70)
plt.xlabel('Estado civil')
plt.ylabel('Nº. de casos')
plt.title('Nº de personas por estado civil')
plt.legend()
plt.show()
```

```python
tcks = df_group_menor.index
tcks_mayor = df_group_mayor.index
plt.bar(tcks-0.25, df_group_menor['number'],
        color='orange', label=' <=50K', width=0.75, align='center', alpha=0.7)
plt.bar(tcks_mayor-0.75, df_group_mayor['number'],
        label=' >50K', width=0.75, align='center', alpha=0.7)
plt.xticks(tcks, labels=df_group_menor['marital-status'], rotation=70)
plt.xlabel('Estado civil')
plt.ylabel('Nº. de casos')
plt.title('Nº de personas por estado civil')
plt.legend()
plt.show()
```

# Múltiples gráficos, diferente eje Y mismo eje X

Mostramos tres gráficos de líneas de diferentes variables relacionadas con el tiempo en el dataframe df2.

El dataframe df2 contiene datos sobre la calidad del aire en una ciudad italiana, con mediciones de diferentes contaminantes y variables meteorológicas.

```python
# crea una lista con los nombres de las columnas que contienen las variables de temperatura,
# humedad relativa y humedad absoluta, respectivamente.

time_cols = ['T', 'RH', 'AH']

# crea una lista con el nombre de la columna que contiene la fecha y la hora de cada medición, que se usará como eje x común para los tres gráficos.
x = ['Date_Time']

# crea una figura con tres subgráficos dispuestos en tres filas y una columna, y asigna los objetos axs a cada subgráfico. El argumento sharex=True indica que se quiere compartir el eje x entre los subgráficos. El argumento figsize especifica el tamaño de la figura en pulgadas.
fig, axs = plt.subplots(len(time_cols), 1, sharex=True, figsize=(14,10))

# Remove horizontal space between axes
# Elimina el espacio horizontal entre los subgráficos, para que se vean más compactos.
fig.subplots_adjust(hspace=0)

#  inicia un bucle que recorre las columnas de la lista time_cols, usando la variable i como índice y la variable col como nombre de la columna.
for i, col in enumerate(time_cols):

    # dibuja un gráfico de línea usando los valores de la columna col del dataframe df2 como eje y, y los valores de la columna x como eje x. El objeto axs[i] se refiere al subgráfico correspondiente al índice i.
    axs[i].plot(df2[x], df2[col])

    # establece la etiqueta del eje x como ‘Fecha’.
    axs[i].set_xlabel('Fecha')

    #establece la etiqueta del eje y como el nombre de la columna correspondiente al índice i.
    axs[i].set_ylabel(time_cols[i])

plt.show()
```

```python
time_cols = ['T', 'RH', 'AH']
colors = ['r', 'g', 'b']
x = ['Date_Time']

fig, axs = plt.subplots(len(time_cols), 1, sharex=True, figsize=(14,10))

# Remove horizontal space between axes
fig.subplots_adjust(hspace=0)

for i, col in enumerate(time_cols):
    axs[i].plot(df2[x],
                df2[col].apply(lambda x: np.nan if x == -200 else x),
                c=colors[i])
    axs[i].set_xlabel('Fecha')
    axs[i].set_ylabel(time_cols[i])

plt.show()
```
